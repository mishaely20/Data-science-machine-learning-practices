# Importing dependency for our lovel Nemo model

```python
import os
import io
import torch
import numpy as np
import pandas as pd
from tqdm import tqdm
import seaborn as sns
import soundfile as sf
import matplotlib.pyplot as plt
from datasets import load_dataset
import nemo.collections.asr as nemo_asr
from datasets import Audio as HF_Audio
from IPython.display import Audio
```

# configaration time bro

```python
class config:
  dataset_path='sartifyllc/Sartify_ITU_Zindi_Testdataset'
  model_name='Koleshjr/kolesh_stt_15_update_mcv13.nemo'
  device='cuda' if torch.cuda.is_available() else 'cpu'
  test_index=100
  batch_size=8
```

loading our datasets

```python
test=load_dataset(config.dataset_path,split='test')
test=test.cast_column('audio',HF_Audio(sampling_rate=16000, decode=False))
test
```

```python
os.makedirs('AelyWaves', exist_ok=True)
audio_paths = [] # Initialize as a list
for i, item in enumerate(tqdm(test)):
  audio = item['audio']['bytes']
  audio_array, sr = sf.read(io.BytesIO(audio))

  # if sr is not 16000 resample it to 16000
  # Resampling code would go here if needed.
  # Example (requires 'resampy' or 'scipy.signal'):
  # if sr != 16000:
  #     audio_array = resampy.resample(audio_array, sr, 16000)
  #     sr = 16000

  audio_path = f'AelyWaves/{i}.wav'
  sf.write(audio_path, audio_array, sr)
  audio_paths.append(audio_path) # Append to the list

# Correcting the print statement to access an element from the list
print(audio_paths[1] if len(audio_paths) > 1 else "Not enough audio files processed to print the second path.")
```

# loading the pretrained model

```python
asr_model=nemo_asr.models.ASRModel.from_pretrained(config.model_name)
```

```python
test_audio=audio_paths[config.test_index]
test_audio
```

```python
transcript=asr_model.transcribe(test_audio)
transcript
```

# Do transcription to all aduios

```python
all_transcripts=asr_model.transcribe(audio_paths,batch_size=config.batch_size)
```

# Prepare submission file

```python
results = []

for i in range(len(test)):
    filename = test[i]['filename']
    text = all_transcripts[i].text
    results.append({'filename': filename, 'text': text})

sub = pd.DataFrame(results)
print(sub.head(2))
```
